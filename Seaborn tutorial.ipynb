```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.displot(tips["total_bill"],kde=True, bins=20)
```

```python
sns.jointplot(x="total_bill",y="size",data=tips,kind='reg')  ##kind='hex'
```

```python
sns.pairplot(data=tips,hue="sex")
```

```python
sns.rugplot(tips['tip']) # it ill work only numerical values
```

```python
sns.kdeplot(tips['total_bill']) # it is also work for numerical values
```

```python
import numpy as np
sns.barplot(x='day',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
tc=tips.corr()
sns.heatmap(tc,annot=True,cmap='magma',linewidth=0.5,linecolor='red')
```
